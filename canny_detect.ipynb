```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to display images
def display_images(images, titles, cmap=None):
    plt.figure(figsize=(15, 8))
    for i, (img, title) in enumerate(zip(images, titles)):
        plt.subplot(2, 4, i + 1)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.imshow(img, cmap=cmap)
        plt.title(title)
        plt.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
# Path to the folder containing images
import os
image_folder = r"D:\Study\Assignment\vision\Canny"  # Replace with your folder path
image_paths = [os.path.join(image_folder, f) for f in os.listdir(image_folder) if f.lower().endswith(('png', 'jpg', 'jpeg'))]
images = [cv2.imread(path) for path in image_paths]

# Ensure images are loaded
if not images:
    print("No valid images found in the folder.")
else:
    print(f"{len(images)} images loaded from folder '{image_folder}'.")

# Part 1: Canny Edge Detection
canny_results = []
for img in images:
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    blurred = cv2.GaussianBlur(gray, (5, 5), 1.4)
    edges = cv2.Canny(blurred, 50, 150)
    canny_results.append(edges)

# Display Original and Canny Edge Detected Images
original_titles = [f"Original {i+1}" for i in range(4)]
canny_titles = [f"Canny {i+1}" for i in range(4)]

print("Displaying Canny Edge Detection Results...")
display_images(images + canny_results, original_titles + canny_titles, cmap='gray')
```

```python
# Path to the folder containing images
import os
image_folder = r"D:\Study\Assignment\vision\equalize"  # Replace with your folder path
image_paths = [os.path.join(image_folder, f) for f in os.listdir(image_folder) if f.lower().endswith(('png', 'jpg', 'jpeg'))]
images = [cv2.imread(path) for path in image_paths]

# Part 2: Histogram Equalization
hist_eq_results = []
for img in images:
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    hist_eq = cv2.equalizeHist(gray)
    hist_eq_results.append(hist_eq)

# Display Original and Histogram Equalized Images
hist_eq_titles = [f"Hist Eq {i+1}" for i in range(4)]

print("Displaying Histogram Equalization Results...")
display_images(images + hist_eq_results, original_titles + hist_eq_titles, cmap='gray')

# Comments on Histogram Equalization (Optional Analysis)
for i, (orig, eq) in enumerate(zip(images, hist_eq_results)):
    print(f"Image {i+1}: Compare original and equalized results visually. Note changes in contrast and detail.")
```
